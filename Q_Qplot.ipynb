```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import pylab
```

```python
measurements=np.random.normal(loc=50,scale=3,size =1000)
print(measurements)
"""loc float or array_like of floats
Mean (“centre”) of the distribution."""
```

```python
plt.plot(measurements)
```

```python
stats.probplot(measurements,dist="norm",plot=pylab)
pylab.show()
#Q-Q plot
```

```python
b=np.random.uniform(low=1,high=2,size=100000)
print(b)
```

```python
stats.probplot(b,dist="norm",plot=pylab)
pylab.show()
#Q-Q plot on uniform varibale vs norm distribution
```
