```python
print("Mean")
print (np.mean(haberman_Long_Survive['Number_of_positive_axillary_nodes_detected']))
print (np.mean(np.append(haberman_Long_Survive['Number_of_positive_axillary_nodes_detected'],50)))
print (np.mean(haberman_Short_Survive['Number_of_positive_axillary_nodes_detected']))
print("stddev")
print(np.mean(haberman_Long_Survive['Number_of_positive_axillary_nodes_detected']))
print(np.mean(haberman_Short_Survive['Number_of_positive_axillary_nodes_detected']))
```
