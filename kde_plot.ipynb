```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
x=stats.norm.rvs(size=1000)
#print(x)
sns.set_style('whitegrid')
sns.kdeplot(np.array(x))
#print(kdeplot)
```

```python
stats.kstest(x,'norm')
```

```python
#since the pvalue is great that is 0.88 this means that the variable x is normally distributed
```

```python
y=np.random.uniform(1,10,10000)
sns.kdeplot(np.array(y),bw=0.9)
print(y)
plt.show()
#bw determines the smoothness of the plotted plot
#kde is a smoothing feature
```

```python
stats.kstest(y,'norm')
```

```python
#since the pvalue is small value then the varible y is not normally distributed
```
